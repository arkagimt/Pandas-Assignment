### Q9. How do you concatenate two Pandas DataFrames?

```python
import pandas as pd
import numpy as np
 
dfa = pd.DataFrame(np.random.randint(25, size=(4, 4)),
                   index=["1", "2", "3", "4"],
                   columns=["A", "B", "C", "D"])
 
dfb = pd.DataFrame(np.random.randint(25, size=(6, 4)),
                   index=["5", "6", "7", "8", "9", "10"],
                   columns=["A", "B", "C", "D"])
 
dfc = pd.DataFrame(np.random.randint(25, size=(4, 4)),
                   columns=["A", "B", "C", "D"])
 
dfd = pd.DataFrame(np.random.randint(25, size=(4, 4)),
                   columns=["E", "F", "G", "H"])
 
display(dfa, dfb, dfc, dfd)
```

```python
# concatenating dfa and dfb along rows
vertical_concat = pd.concat([dfa, dfb], axis=0)
 
# concatenating dfc and dfd along columns
horizontal_concat = pd.concat([dfc, dfd], axis=1)

display(vertical_concat, horizontal_concat)
```

### Q15. How do you create a copy of a Pandas DataFrame?

```python
data = [['dom', 10], ['abhi', 15], ['celeste', 14]]
df = pd.DataFrame(data, columns = ['Name', 'Age'])

df_copy = df.copy(deep=True)

print("Original Dataframe - \n")
print(df)

print("Deep Copy Dataframe - \n")
print(df_copy)

print("\n")
print("Changing value in original dataframe\n")
df.iloc[0,1] = -9

print("Original Dataframe after changes - \n")
print(df)

print("Deep Copy Dataframe after changes - \n")
print(df_copy)
```

### Q23. How do you create a new column in a DataFrame based on the values of another column?

```python
# Creating the DataFrame
df = pd.DataFrame({'Date':['10/2/2011', '11/2/2011', '12/2/2011', '13/2/2011'],
                    'Event':['Music', 'Poetry', 'Theatre', 'Comedy'],
                    'Cost':[10000, 5000, 15000, 2000]})
 
# Print the dataframe
df
```

```python
# using apply function to create a new column
df['Discounted_Price'] = df.apply(lambda row: row.Cost -
                                  (row.Cost * 0.1), axis = 1)
df
```

###### OR

```python
df = pd.DataFrame({'Date':['10/2/2011', '11/2/2011', '12/2/2011', '13/2/2011'],
                    'Event':['Music', 'Poetry', 'Theatre', 'Comedy'],
                    'Cost':[10000, 5000, 15000, 2000]})
 
# Create a new column 'Discounted_Price' after applying
# 10% discount on the existing 'Cost' column.
 
# create a new column
df['Discounted_Price'] = df['Cost'] - (0.1 * df['Cost'])
 
# Print the DataFrame after
# addition of new column
df
```

###### OR using DataFrame.map()

```python
data = {
    "name": ["John", "Ted", "Dev", "Brad", "Rex", "Smith", "Samuel", "David"],
    "salary": [10000, 20000, 50000, 45500, 19800, 95000, 5000, 50000]
}
# create dataframe from data dictionary
df = pd.DataFrame(data)
# print the dataframe
df.head()
```

```python
def salary_stats(value):
    if value < 10000:
        return "very low"
    if 10000 <= value < 25000:
        return "low"
    elif 25000 <= value < 40000:
        return "average"
    elif 40000 <= value < 50000:
        return "better"
    elif value >= 50000:
        return "very good"
 
df['salary_stats'] = df['salary'].map(salary_stats)
display(df.head())
```

### Q25. What is the difference between .loc and .iloc in Pandas?

## loc in Pandas
###### 1. Label-based data selector
###### 2. Indices should be sorted in order, or loc[ ] will only select the mentioned indices when slicing
###### 3. Indices should be numerical, else slicing cannot be done
###### 4. The end index is included during slicing
###### 5. Accepts bool series or list in conditions

```python
df = pd.DataFrame({'team': ['A', 'A', 'A', 'A', 'B', 'B', 'B', 'B'],
                   'points': [5, 7, 7, 9, 12, 9, 9, 4],
                   'assists': [11, 8, 10, 6, 6, 5, 9, 12]},
                   index=['A', 'B', 'C', 'D', 'E', 'F', 'G', 'H'])

#view DataFrame
df
```

```python
#select rows with index labels 'E' and 'F'
df.loc[['E', 'F']]
```

```python
#select 'E' and 'F' rows and 'team' and 'assists' columns
df.loc[['E', 'F'], ['team', 'assists']]
```

### liloc in Pandas
##### 1. Index-based data selector
##### 2. Indices need not be sorted in order when slicing
##### 3. Indices can be numerical or categorical
##### 4. The end index is excluded during slicing
##### 5. Only accepts bool list in conditions

```python
#create DataFrame
df = pd.DataFrame({'team': ['A', 'A', 'A', 'A', 'B', 'B', 'B', 'B'],
                   'points': [5, 7, 7, 9, 12, 9, 9, 4],
                   'assists': [11, 8, 10, 6, 6, 5, 9, 12]},
                   index=['A', 'B', 'C', 'D', 'E', 'F', 'G', 'H'])

#view DataFrame
df
```

```python
#select rows in index positions 4 through 6 (not including 6)
df.iloc[4:6]
```

```python
#select rows in range 4 through 6 and columns in range 0 through 2
df.iloc[4:6, 0:2]
```
